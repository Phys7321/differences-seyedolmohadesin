```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot
import math
from math import sin, cos, pi

import myder
#from myder import higherdiff, seconddiff
from array import array
from numpy import array
```

```python

def higherdiff(f,a,b,N):
    h = (b-a)/N
    g=np.zeros(N)
    for k in range(0,N):
        slop = ((1/24)*f(a+(k-3/2)*h)-(27/24)*f(a+(k-1/2)*h)+(27/24)*f(a+(k+1/2)*h)-(1/24)*f(a+(k+3/2)*h))/h
        g[k]=slop
    return g

def seconddiff(f,a,b,N):
    h = (b-a)/N
    l=np.zeros(N)
    for k in range(0,N):
        slop = (f(a+(k+1)*h)-2*f(a+k*h)+f(a+(k-1)*h))/h**2
        l[k]=slop
    return l


```

```python
# defining function y as $\sin^2{\frac{1}{x(2-x)}}$ 

x=np.linspace(0.01,1.99,num=100)
y=[]

f=lambda x: sin(1/(x*(2-x)))**2

for i in x:
    y.append(f(i))
    
    
#first derivative with higher order method : degree 3

N=100
a=0.01
b=1.99
h = (b-a)/N
 
yp=higherdiff(f,a,b,N) # first derivative 


    
# first derivative functional form 

fp=lambda x: sin(2/(x*(2-x)))*2*(x-1)/((x*(2-x))**2)
yp_function=list(map(fp,x))

#second derivative (numerically)

ypp=seconddiff(f,a,b,N) # second derivative


```
